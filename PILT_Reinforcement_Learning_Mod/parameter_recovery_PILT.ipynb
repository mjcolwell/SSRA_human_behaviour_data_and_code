```python
def rescorla_wagner(Y, alpha, start):
    """
    Y: array of outcomes
    alpha: learning rate
    start: initial value estimate
    Returns an array of value estimates.
    """
    r = np.zeros(len(Y))
    r[0] = start
    for i in range(1, len(Y)):
        r[i] = r[i-1] + alpha * (Y[i-1] - r[i-1])
    return r
```

```python
def simulate_choice_rho(type, schedule, permutation, num_perms):
    track_lr =[]
    track_rho = []
    track_choice = []
    track_prob = []
    for p in permutation: 
        lr = p[0]
        rho = p[1]
        # Generate win and loss values for stimuli A, using the outcomes from `schedule`
        starting_val = 0.5
        schedule_modified = [s * rho for s in schedule]
        vals = rescorla_wagner(schedule_modified, lr, starting_val)
        bel = vals - starting_val # Adjusted for a simplified scenario
        # Calculate choice probability
        choice_prob = 1 / (1 + np.exp(-bel))
        
        # Run random bit repeatedly to get distribution
        for _ in range(num_perms):
            # Generate choices
            dice = np.random.rand(len(choice_prob))
            choice = choice_prob > dice  # 1 when option 1 chosen, 0 when alternative option is chosen
            track_choice.extend(choice)
            track_prob.extend(choice_prob)
            track_lr.extend([lr]*len(schedule))
            track_rho.extend([rho]*len(schedule))
            
    df_simulated = pd.DataFrame({"simulation_lr": track_lr,
                                  "simulation_rho": track_rho,
                                  "simulated_choice": track_choice,
                                "choice_prob": track_prob})
    return df_simulated

```
